```python
def plot_boxplot(df,ft):
    df.boxplot(column=[ft])
    plt.grid(False)
    plt.show()
    
```

```python
plot_boxplot(df_raw,"marketCap")
```

```python
def outlier(df,ft):
    A = df[ft].quantile(0.25)
    B = df[ft].quantile(0.75)
    IQR=B-A
    
    lower_bound = B - 1.5 * IQR
    upper_bound = B + 1.5 * IQR
    
    lst = df.index[(df[ft] < lower_bound) | (df[ft] > upper_bound) ]
    return lst
```

```python
index_list = []
for features in ['marketCap', 'volume']:
    index_list.extend(outlier(df_raw,features))
```

```python
index_list
```

```python
def remove(df, lst):
    ls = sorted(set(lst))
    df= df.drop(lst)
    return df
```

```python
df_cleaned = remove(df_raw, index_list)
```

```python
df_cleaned.shape
```

```python
plot_boxplot(df_cleaned, 'marketCap')
```

```python
plot_boxplot(df_cleaned, 'volume')
```
